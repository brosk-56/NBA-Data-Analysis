```python
# BAR CHART
import matplotlib.pyplot as plt
import pandas as pd

# DataFrame `df_csv` with 'Height' and 'Salary' columns
# Calculate average salary by height group
avg_salary = df_csv.groupby('Height')['Salary'].mean()

plt.figure(figsize=(7, 3))  # deciding on sizes of the chart

# Plotting the BAR CHART
avg_salary.plot(kind='bar')

# Print the result (average salary for each height group)
print(avg_salary)

plt.title('Histogram of average salary for each height group')  # title of the chart displayed
plt.xlabel('Height')  # choosing x zone
plt.ylabel('Salary')  # choosing y zone

# Display the chart
plt.show()
```

```python
#Average salary by height in the pie chart in the first chart and in the second one proportion of height against the total height
import matplotlib.pyplot as plt
import pandas as pd

# DataFrame `df_csv` with 'Height' and 'Salary' columns
# Calculate average salary by height group
avg_salary = df_csv.groupby('Height')['Salary'].mean()
height_counts = df_csv['Height'].value_counts()

# Plotting Pie Chart - Average Salary by Height
plt.figure(figsize=(7, 4))  # Setting the size of the figure
# Plotting the pie chart and displaying percentages on the slices
# The percentages represent each height group's average salary as a proportion of the sum of average salaries
avg_salary.plot(kind='pie', autopct='%1.1f%%')
plt.title('Average Salary for Each Height Group')  # Adding a title to the chart
plt.ylabel('')  # Hiding the 'Height' label as it's redundant in a pie chart
plt.show()

# Plotting Pie Chart - Height Proportion
plt.figure(figsize=(7, 4))  # Setting the size of the figure
# Plotting the pie chart and displaying percentages on the slices
# The percentages represent the proportion of each height group relative to the total number of entries
height_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Each Height Group')  # Adding a title to the chart
plt.ylabel('')  # Hiding the 'Height' label as it's redundant in a pie chart
plt.show()

```

```python
# Maximum salary by height in the pie chart
import matplotlib.pyplot as plt
import pandas as pd

# DataFrame `df_csv` with 'Height' and 'Salary' columns
# Calculate maximum salary by height group
max_salary = df_csv.groupby('Height')['Salary'].max()

# Plotting Pie Chart - Maximum Salary by Height
plt.figure(figsize=(7, 4))  # Setting the size of the figure
# Plotting the pie chart and displaying percentages on the slices
# The percentages represent each height group's maximum salary as a proportion of the sum of maximum salaries
max_salary.plot(kind='pie', autopct='%1.1f%%')
plt.title('Maximum Salary for Each Height Group')  # Adding a title to the chart
plt.ylabel('')  # Hiding the 'Height' label as it's redundant in a pie chart
plt.show()
```

```python
# Finding out the top 20 teams with a maximum salary
import matplotlib.pyplot as plt
import pandas as pd
df_csv = pd.read_csv(r"C:\Users\bulen\Downloads\nba.csv")

#top 20 most paid salary by teams from largest 
ave_salary = df_csv.groupby('Team')['Salary'].max().nlargest(20)
plt.title('Maximum Salary by Teams')
ave_salary.plot(kind='bar', color='b')
plt.xlabel('Teams')
plt.ylabel('Maximum Salary')
plt.show()
```

```python
# Bar chart with top 20 teams by average salary
import matplotlib.pyplot as plt
import pandas as pd
df_csv = pd.read_csv(r"C:\Users\bulen\Downloads\nba.csv")

#top 20 average salary by teams 
ave_salary = df_csv.groupby('Team')['Salary'].mean().head(20)
plt.title('Average Salary by Teams')

ave_salary.plot(kind='bar', color='b')
plt.xlabel('Teams')
plt.ylabel('Average Salary')
plt.show()
```

```python
# Pie chart with teams by average age, the higher the percentage older the team
import matplotlib.pyplot as plt
import pandas as pd
df_csv = pd.read_csv(r"C:\Users\bulen\Downloads\nba.csv")

ave_age = df_csv.groupby('Team')['Age'].mean().nlargest(20)
plt.figure(figsize=(6,6))
plt.title('Average Age by Teams')

#average ages by teams in the clockwise in the pie
ave_age.plot(kind='pie',autopct='%1.1f%%',counterclock=True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# In this code, the reset_index() method is used to convert the Series into a DataFrame, enabling the scatter plot.
# Calculate average salary by height and reset the index to create a DataFrame
avg_salary = df_csv.groupby('Height')['Salary'].mean().reset_index()

# Plot scatter plot
avg_salary.plot(kind='scatter', x='Height', y='Salary')
plt.grid()
# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your DataFrame
df_csv = pd.read_csv(r"C:\Users\bulen\Downloads\nba.csv")

# Convert the 'Weight' column to numeric, forcing non-numeric values to NaN
df_csv['Weight'] = pd.to_numeric(df_csv['Weight'], errors='coerce')

# Drop rows with NaN values in the 'Weight' column
df_csv.dropna(subset=['Weight'], inplace=True)

# Calculate the average weight by team- top 20 team
ave_weight = df_csv.groupby('Team')['Weight'].mean().head(20)

plt.figure(figsize=(6,4))

# Plotting the bar chart
ave_weight.plot(kind='bar', color='blue')

plt.title('Average Weight grouped by Team')
plt.xlabel('Team')
plt.ylabel('Weight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your DataFrame
df_csv = pd.read_csv(r"C:\Users\bulen\Downloads\nba.csv")
df_csv['Weight'] = pd.to_numeric(df_csv['Weight'],errors='coerce') # converting Weight to numeric if any non-numeric in the column
df_csv.dropna(subset=['Age','Weight'],inplace=True)   #WE HAVE TO USE SUBSET WITH DROPNA TOGETHER DROPS NANS IN ROWS WITH SPECIAL COLUMNS
average_weight = df_csv.groupby('Age')['Weight'].mean().nlargest(20)
average_weight.plot(kind = 'bar',color='red')# WHILE DROPNA DROPS ALL NANS WHETHER GOOD OR BAD DATA    
plt.title('Average weight grouped by age')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.show()
```
